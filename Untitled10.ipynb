```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
df= pd.read_excel("microretailer_mit_lift_lab.xlsx")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.fillna(method="ffill")
df=df.fillna(method="bfill")
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
def remove_outliers_iqr(df1, column_name, threshold=1.5):
    Q1 = df1[column_name].quantile(0.25)
    Q3 = df1[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    df1 = df1[(df1[column_name] >= lower_bound) & (df1[column_name] <= upper_bound)]
    return df1


data_clean = pd.concat([df, df.select_dtypes(exclude=['object'])], axis=1)

data_clean.shape
```

```python
table1= freq_tbl(data_clean['_title'])
table1
```

```python
import matplotlib.pyplot as plt
Filtro= table1[table1['frequency']>1]
Filtro_index= Filtro.set_index('_title')
Filtro_index.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Herramientas de relación con los clientes')
plt.xlabel('Tipo de Herramienta')
plt.ylabel('Frecuencia')
```

```python
Filtro_index.plot(kind='area', figsize=(10,4),alpha = 0.8)
```
